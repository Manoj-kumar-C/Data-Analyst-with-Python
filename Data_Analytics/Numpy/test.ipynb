```python

%pip install numpy
import numpy as np

# Arrays
arr = np.array([1, 2, 3, 4, 5])
print("Array:", arr)

# Array Indexing
print("First element:", arr[0])
print("Last element:", arr[-1])

# Array Operations
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
print("Addition:", arr1 + arr2)
print("Multiplication:", arr1 * arr2)

# Broadcasting
arr3 = np.array([[1], [2], [3]])
arr4 = np.array([4, 5, 6])
print("Broadcasting:", arr3 + arr4)

# Array Manipulation
arr5 = np.array([[1, 2, 3], [4, 5, 6]])
print("Reshaping:", arr5.reshape(3, 2))
arr6 = np.array([[7, 8, 9]])
print("Concatenation:", np.concatenate((arr5, arr6), axis=0))

# Universal Functions (ufuncs)
print("Sine function:", np.sin(arr1))

# Vectorization
arr9 = np.array([1, 2, 3, 4])
print("Vectorized operation:", arr9 * 2)

# Random Number Generation
rand_arr = np.random.rand(3, 3)
print("Random numbers:", rand_arr)

# File Input/Output
np.save('my_array.npy', arr1)
loaded_arr = np.load('my_array.npy')
print("Loaded array:", loaded_arr)
```
